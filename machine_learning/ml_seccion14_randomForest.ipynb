# Random Forest - Clasificación


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

vinos = pd.read_csv('recursos/vino.csv')
vinos
```

```python
# separar entre caracteristicas y tipo de vino
X = vinos.drop('Wine Type', axis=1)
y = vinos['Wine Type']

```

```python
# dividimos en datos de entrenamiento y datos de prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# entrenar en modelo y hacer las predicciones

from sklearn.ensemble import RandomForestClassifier
```

```python
# numero de estimadores 80
# creamos el modelo
random_forest = RandomForestClassifier(n_estimators=80)

# entrenamos el modelo
random_forest.fit(X_train, y_train)
```

```python
# realizamos las predicciones
predicciones = random_forest.predict(X_test)
```

```python
# comparamos las predicciones con el y_test
predicciones
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predicciones))
# el 1.00 significa que la precisión es del 100%
```

```python
# imprimimos la matriz de confusion
print(confusion_matrix(y_test, predicciones))
```
