```python
#matplotlib es una biblioteca para la generación de graficos a partir de dataframes
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2,9,10)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y, 'red') #red o r sería valido tmbn
plt.title('Este es mi grafico') #titulo del grafico
plt.xlabel('Eje x') #poner el texto a los ejes
plt.ylabel('Eje y')
plt.show()
```

```python
#1 fila dos columnas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'blue')
```

```python
#2 fila 1 columna
plt.subplot(2, 1, 1)
plt.plot(x, y, 'g')

plt.subplot(2, 1, 2)
plt.plot(y, x, 'blue')
```

Programación orientada a objetos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1, 0.1, 0.5, 0.5]) #espacio q dejamos x debajo y x la izqda y el ancho y el alto
grafico.plot(x, y, 'green')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1, 0.1, 0.9, 0.9]) #espacio q dejamos x debajo y x la izqda y el ancho y el alto
grafico1.plot(x, y, 'r')

grafico2 = figura.add_axes([0.2, 0.5, 0.3, 0.3]) #espacio q dejamos x debajo y x la izqda y el ancho y el alto
grafico2.plot(x, y, 'blue')
```

Multigráficos (filas, columnas)

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)
graficos[0][0].plot(x,y,'blue')
graficos[0][1].plot(x,y,'r')
graficos[1][0].plot(x,y,'green')
graficos[1][1].plot(x,y,'yellow')

graficos[0][0].set_title('azul')
graficos[0][1].set_title('rojo')
graficos[1][0].set_title('verde')
graficos[1][1].set_title('amarillo')
```

Tamaño del gráfico

```python
figura = plt.figure(figsize=(8,2))
grafico = figura.add_axes([0.1, 0.1, 0.9, 0.9])
grafico.plot(x, y, 'r')
```

**Dos gráficos en la misma figura**

```python
figura = plt.figure()
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, label='Al cuadrado')
graficos.plot(x, x**3, label='Al cubo')

#añadir una leyenda
graficos.legend(loc=0)
```

**Color del gráfico**

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,0.5,0.5])
grafico.plot(x,y, color='#0DED14') # tambien se pueden poner los colores hexadecimales o el nombre del color color='tomato'
```

**linea del gráfico (grosor, transparencia y tipo)**

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,0.5,0.5])
grafico.plot(x,y, color='tomato', linewidth=5, alpha=0.5, linestyle=':') #alpha para transparencia, 0.5 es lo medio, linestyle para el tipo de linea
#pej '-' es la lin normal, '-.' para punto raya
```

## Plotly y Cufflinks -- Visualizacioes interactivas

```python
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
np.str_ = np.str_
```

```python
init_notebook_mode(connected=True)
cf.go_offline()
%matplotlib inline

dataframe = pd.DataFrame(np.random.randn(100,4), columns=['a','b','c','d'])
dataframe
```

```python
dataframe.plot()
```

```python
#grafico interactivo
dataframe.iplot()
```

```python
dataframe.iplot(kind='scatter', x='a', y='b', mode='markers')
```

```python
dataframe.iplot(kind='bar')
```

```python
#crea un diagrama de barras con la suma de cada columna
dataframe.sum().iplot(kind='bar')
```

```python
dataframe.iplot(kind='box')
```

```python
dataframe['a'].iplot(kind='hist', bins=30)
```

```python
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
cf.go_offline()
%matplotlib inline

dataframe = pd.DataFrame(np.random.randn(100,4), columns=['a','b','c','d'])


dataframe[['a','b']].iplot(kind='spread')
```

```python
dataframe.iplot(kind='bubble', x='a', y='b', size='c')
```
