```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

vinos = pd.read_csv('C:/Users/SARAPC/Dropbox/phyton/machine learning/recursos/vino.csv')
vinos.head(10)
```

```python
vinos['Wine Type'].unique()
```

```python
# podemos saber la cantidad de cada tipo
vinos['Wine Type'].value_counts()
```

```python
X = vinos.drop('Wine Type', axis=1)
y = vinos['Wine Type']
X

```

```python
y
```

```python
# dividimos en datos de entrenamiento y datos de prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier

arbol = DecisionTreeClassifier()

#entrenamos el arbol
arbol.fit(X_train, y_train)
```

```python
# creamos una variable de predicciones
predicciones = arbol.predict(X_test)
predicciones
```

```python
#valores reales de las pruebas
y_test
```

```python
#evaluar el modelo
#primero generar un informe para ver la precisión del modelo y por otro lado la matriz de confusión para ver cuantos falsos positivos y falsos negativos hay
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#esto nos evalua el modelo
print(classification_report(y_test, predicciones))
```

```python
# positivos correctos y falsos positivos
print(confusion_matrix(y_test, predicciones))
```
